## 2. Implementing Gradient Descent Algorithm (60 pts)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x).
def f(x):
    return np.sqrt(x**2 + 5)

# Plot the function f(x), np.linspace might be helpful.
x_vals = np.linspace(-5, 5, 500)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r"$f(x) = \sqrt{x^2 + 5}$")
plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Define the derivative of the function f(x), for gradient descent
def f_prime(x):
    return x / np.sqrt(x**2 + 5)

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradient_descent(x_0, step_size, num_iterations):
    x_values = [x_0]
    for _ in range(num_iterations):
        x_new = x_values[-1] - step_size * f_prime(x_values[-1])
        x_values.append(x_new)
    return x_values

# set up/assign initial values
last_digit = 3 #SBU ID - 114980843
x_0 = last_digit / 5 + 1  # Initial guess
step_sizes = [5, 3, 1, 0.5]  # Different step sizes
num_iterations = 50  # Number of iterations

# Run gradient descent by calling the function
x_k_values = gradient_descent(x_0, step_size=1, num_iterations=num_iterations)

# report x_T by printing out the value
x_T = x_k_values[-1]
print(f"The final value of x_T for step size 1 is {x_T:.4f}")

# Plot the f(x_k) values vs. iteration number
f_xk_values = [f(x) for x in x_k_values]

plt.figure(figsize=(8, 5))
plt.plot(range(len(f_xk_values)), f_xk_values, label="Step Size = 1")
plt.title(r"$f(x_k)$ vs. Iteration Number")
plt.xlabel("Iteration Number (k)")
plt.ylabel(r"$f(x_k)$")
plt.legend()
plt.grid()
plt.show()
```

```python
# create a list containing step sizes
step_sizes = [5, 3, 1, 0.5]

# initialize a list to store the results from different step sizes
results = {}

# loop through the step sizes and perform GD and record results
for step_size in step_sizes:
    results[step_size] = gradient_descent(x_0, step_size, num_iterations)

# plot multiple curves in a single graph
plt.figure(figsize=(8, 5))
for step_size in step_sizes:
    x_k_values = results[step_size]
    f_differences = [f(x_k_values[i - 1]) - f(x_k_values[i]) for i in range(1, len(x_k_values))]
    plt.plot(range(len(f_differences)), f_differences, label=f"Step Size = {step_size}")

plt.title(r"$f(x_{k-1}) - f(x_k)$ vs. Iteration Number for Different Step Sizes")
plt.xlabel("Iteration Number (k)")
plt.ylabel(r"$f(x_{k-1}) - f(x_k)$")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Generate synthetic data
np.random.seed(43)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias term (intercept) in the normal equation
X_b = np.c_[np.ones((100, 1)), X]  # Add a column of ones to X

# Find theta (coefficients) using the normal equation
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# Print the intercept and coefficients of the regression plane
intercept = theta[0]
coefficients = theta[1:]
print(f"Intercept: {intercept}")
print(f"Coefficients: {coefficients}")

# Plot the generated data and the regression plane
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the original data
scatter = ax.scatter(X[:, 0], X[:, 1], y, color='b', label='Original Data')

# Create a meshgrid for the regression plane
x1 = np.linspace(0, 2, 10)
x2 = np.linspace(0, 2, 10)
x1, x2 = np.meshgrid(x1, x2)
y_plane = intercept + coefficients[0] * x1 + coefficients[1] * x2

# Plot the regression plane
plane = ax.plot_surface(x1, x2, y_plane, alpha=0.5, color='yellow')

# Manually create a legend
scatter_proxy = plt.Line2D([0], [0], linestyle="none", marker="o", color='b', label="Original Data")
plane_proxy = plt.Line2D([0], [0], linestyle="none", marker="s", markersize=10, color='yellow', label="Regression Plane")
ax.legend(handles=[scatter_proxy, plane_proxy])

# Labeling axes
ax.set_title("Linear Regression Plane")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("y")
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.
2. Code Explanation: To enhance code comprehension, please augment the code with meaningful comments. These comments should elucidate the purpose and functionality of each code segment, making it easier for readers to understand the logistic regression implementation.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function: Converts linear values into probabilities between 0 and 1
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Set a random seed for reproducibility
np.random.seed(595)

# Generate random data points (2 features, 100 samples)
X = np.random.rand(2, 100)

# True weights and bias for generating synthetic labels
w_true = np.array([1.5, -2.5])
b_true = 1.0  # True bias

# Generate probabilities using the sigmoid function
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)

# Generate binary labels based on probabilities
Y = (probabilities > 0.5).astype(int)

# Add noise to the features
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test = X[:, :80], X[:, 80:]
Y_train, Y_test = Y[:80], Y[80:]

# Initialize weights and bias
w = np.zeros(X.shape[0])  # Initialize weights to zero
b = 0.0  # Initialize bias to zero

# learning rate and number of epochs
learning_rate = 0.1  # Adjusted for stable convergence
num_epochs = 500  # Sufficient for convergence

# Gradient Descent: Iteratively update weights and bias
for epoch in range(num_epochs):
    # Compute predictions using the current weights and bias
    A_train = sigmoid(np.dot(w.T, X_train) + b)

    # Compute gradients for weights and bias
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    dJdb = np.mean(A_train - Y_train)

    # Update weights and bias using gradients and learning rate
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# Final predictions for training data
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

# Final predictions for testing data
A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# Compute accuracy for training and testing sets
train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

# Print accuracy scores
print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
# Plot decision boundary for training set
plt.scatter(X_train[0, :], X_train[1, :], c=Y_train, cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

# Decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary (Training Set)')
plt.show()
```

```python
# Plot decision boundary for testing set
plt.scatter(X_test[0, :], X_test[1, :], c=Y_test, cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

# Decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary (Testing Set)')
plt.show()
```
